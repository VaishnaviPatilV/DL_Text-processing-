```python
# punctuation

#"!#$%^&*#&(&^)"
import string, time
string.punctuation
```

```python
exclude = string.punctuation
```

```python
def remove_punc(text):
  for i in exclude:
    text = text.replace(i, "")
  return text
```

```python
# Chat Words Removal
# LOL , IDK, BRB, OMG --- how ml will understand this so what we will do will parse the dictionary so that
# ML can understand this
```

```python
chat_words = {
    'AFAIK' : 'As Far As I Know',
    'AFK' : 'Away From Keyboard',
    'ASAP' : 'As Soon As Possible',
    'FYI' : 'For Your Information',
    'ASAP': 'As Soon As Possible',
    'BRB' : 'Be Right Back',
    'TTYL': 'Talk To You Later',
    'IDK' : "I Don't Know",
    'GTG' : 'Got To Go',
    'TMI' : 'Too Much Information',
    'LOL' : 'Laugh Out Loud',
    'BTW' : 'By The Way',
    'ICYMI': 'In Case You Missed It',
    'FAQ' : 'Frequently Asked Question',
    'TGIF': "Thank God It's Friday",
    'IMO' : 'In My Opinion'
}
```

```python
def chat_conv(text):
  new_text = []
  for w in text.split():
    if w.upper() in chat_words:
      new_text.append(chat_words[w.upper()])
    else:
      new_text.append(w)
  return " ".join(new_text)
```

```python
chat_conv("Do this ASAP")
chat_conv("Aaj humara lol hogya")
```

```python
from nltk.corpus import stopwords

# Load stopwords once and convert to set for fast lookup
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    # keep only words not in stopwords
    return " ".join([w for w in text.split() if w.lower() not in stop_words])
```

```python
remove_stopwords("I had everything when I was small, all the friends all my toys, but now I have nothing !!!")
```

```python
df['review'] = df['review'].apply(remove_stopwords)
```
